```python
from scipy.constants import zero_Celsius, R
import matplotlib.pyplot as plt
import numpy as np
```

```python
T1 = zero_Celsius + 10
T2 = zero_Celsius + 40

V = np.linspace(0.07,0.3, 1000)*0.001

a= 0.37
b= 4.3e-5

def p_calc(V,T):
    return R*T/(V-b)-a/(V**2)
```

```python
plt.plot(V,p_calc(V,T1), label = "T1")
plt.plot(V,p_calc(V,T2), label = "T2")
plt.legend()
plt.xlabel("Volume in m$^3$")
plt.ylabel("Pressure in Pa");
```
