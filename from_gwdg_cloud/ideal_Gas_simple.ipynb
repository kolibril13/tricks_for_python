```python
a= 0.37
b= 4.3e-5
```

```python
from scipy.constants import zero_Celsius,R
```

```python
T1 = zero_Celsius + 10
T2 = zero_Celsius + 40
```

```python
import numpy as np
```

```python
V = np.linspace(0.07,0.3, 1000)*0.001
```

```python
def p_calc(V,T):
    return R*T/(V-b)-a/(V**2)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"]= 200
```

```python
plt.plot(V,p_calc(V,T1), label = "T1")
plt.plot(V,p_calc(V,T2), label = "T2")
T3 = zero_Celsius+ 5
plt.plot(V,p_calc(V,T3), label = "T3")

plt.legend()
plt.xlabel("Volume in m$^3$")
plt.ylabel("Pressure in Pa")
```
