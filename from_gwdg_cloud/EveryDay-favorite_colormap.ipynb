```python

import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler
plt.rcParams['figure.dpi'] = 60  
plt.rcParams['axes.prop_cycle'] = cycler(color = ['mediumspringgreen','m','skyblue','dodgerblue','navy'])

def gaussian(x, mu, sig):
    return 1/(sig*np.sqrt(2*np.pi))*np.exp( ( -1/2 * ( (x-mu)/sig)**2 ) )
x_values = np.linspace(-5, 5, 220)
mu=0; sig= 1

plt.figure(figsize=(9, 5))
plt.plot(x_values, gaussian(x_values, mu, sig) )
plt.plot(x_values, gaussian(x_values, mu+0.1, sig) )
plt.plot(x_values, gaussian(x_values, mu+0.2, sig) )
plt.plot(x_values, gaussian(x_values, mu+0.3, sig) )
plt.plot(x_values, gaussian(x_values, mu+0.4, sig) )
```
