```python
import skimage
from skimage import data,io
import numpy as np
import matplotlib.pyplot as plt
```

```python
from PIL import Image
def iDisp(im,scale=1):
    newsize=[int(im.shape[1]*scale), int(im.shape[0]*scale)]
    if len(im.shape)==3:
        im=im[...,::-1]
    return Image.fromarray(im.astype(np.uint8)).resize(newsize, Image.NEAREST).convert("RGB")
```

```python
camera = data.camera()
camera.shape
```

```python
iDisp(camera)
```

```python
img = plt.imshow(camera, interpolation='nearest')
img.set_cmap('gray')
plt.axis('off');
```
