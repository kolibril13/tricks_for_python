## Linear Step Function
$$
f(x) =  
\begin{cases} 0,           & x < x_0 
           \\ \frac{(x)}{x_1-x_0}+ \frac{x_0}{x_0-x_1}, & x_0 < x < x_1
           \\ 1,           & x_1 < x 
\end{cases}
$$

```python
x= np.linspace(0., 10., 100)
def linear_step_func(x,x0,x1):
    y= np.piecewise(x, [
        x < x0, 
       (x >= x0) & (x <= x1), 
        x > x1],
            [0.,
            lambda x: x/(x1-x0)+x0/(x0-x1), 
             1.]
       )
    return y
plt.plot(x, linear_step_func(x,2, 5))
plt.show()
```

## Heaviside-Step function
$$
f(x) =  
\begin{cases} 0,           & x < 0 ,
           \\ z,           & x=0
           \\ 1,           & x > 0
\end{cases}
$$

```python
x= np.linspace(-10, 10., 100)
z=0
y=np.heaviside(x,z) # second argument: f(x=0)=0
plt.plot(x, y);
```

```python
t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))
plt.figure(figsize=(9, 5))
plt.axhline(c="k")
plt.axvline(c="k")
plt.axhline(y=0.5,c="k", ls = ":")
plt.axhline(y=1.0,c="k", ls = ":")
plt.plot(t, sig, "b-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
plt.xlabel("t")
plt.legend(loc="upper left", fontsize=20)
plt.show()
```

##  $$ \psi(x, t) = \sum \limits_j C_j \cdot e^{\mathrm{i}( - k_j \cdot x + \omega_j \cdot t )}, \omega_j= \frac{k_j}{c(k_j)} $$ 
## $$  \Rightarrow \psi(x, t) = \sum_j C_j \cdot e^{-i( k_j \cdot x -  \frac{k_j}{c(k_j)} \cdot t )}, \ $$ 

```python
# implementation 1
x = np.linspace(-100,100,400)
def g(x,k_size):
    k_all=np.linspace(0,0.5,k_size)
    x_m, k_m = np.meshgrid(x,k_all)
    matrix= np.exp(-1j*(k_m*x_m - (k_m/(k_m**1.1+1))*50) )
    my_sum=np.sum(matrix,axis=0)
    return my_sum
plt.axhline(c="k")
plt.axvline(c="k")
plt.plot(x,np.real(g(x,k_size=100)), label = r"$\psi(x, t) = \sum_j C_j \cdot e^{-i( k_j \cdot x -  \frac{k_j}{c(k_j)} \cdot t )} \ $");
plt.legend(loc='upper left')
```
